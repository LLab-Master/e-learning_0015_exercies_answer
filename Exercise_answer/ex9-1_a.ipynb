```python
import numpy as np
Z = np.load('vectorize_result_all3.npz',allow_pickle=True)
X = Z['arr_0']
print(len(X[0]))
labels = np.load('labels3.npy',allow_pickle=True).astype('float32')
fnames = np.load('fnames3.npy',allow_pickle=True)
```

```python
#次元削減(主成分分析)

from sklearn.decomposition import PCA
pca = PCA(n_components=300)
G = pca.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# 学習用とテスト用に分離する 
x_train, x_test, y_train, y_test = train_test_split(G, labels, test_size = 0.2, train_size = 0.8, shuffle = True)
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf', C=100, gamma=0.0001)

clf.fit(x_train, y_train)
pred = clf.predict(x_test)
score = accuracy_score(y_test, pred)
print("正解率 = " , score)
```

テストデータを分類して、結果を確認してみる。

```python
pred = clf.predict(x_test)
```

```python
LABEL_NAME = ['dokujo-tsushin', 'it-life-hack','kaden-channel', 
             'livedoor-homme', 'movie-enter', 'peachy','smax',
             'sports-watch', 'topic-news']
```

```python
# ファイルの最初だけ返す
def file_contents(fname):
    with open(fname, 'r') as f:
        contents = ''.join(f.readlines()[2:])
    return contents[:50]   # 最初の50文字だけ
```

```python
ok = 0
ng = 0

for i, p in enumerate(pred):
    if p != y_test[i]:
        ng += 1
        # 予測を間違えたものだけ表示
        print(fnames[i])
        print('True:', LABEL_NAME[int(labels[i])], '  Predicted:', LABEL_NAME[int(p)])
        print(file_contents(fnames[i]))
        print()
    else:
        ok += 1

print("OK:", ok, "  NG:", ng, " Accuracy:{:.4}".format(ok/(ok + ng)))
```
