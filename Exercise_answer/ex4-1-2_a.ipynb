```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np

from sklearn import datasets
```

```python
### データ読み込み
boston = datasets.load_boston()

### データの様子を確認
print(boston.data[:, 0])
print(boston.data[:, 5])
```

### 次のセルに演習4-1の解答を記述してください

```python
# 演習4-1の解答欄

from sklearn import preprocessing

for i in range(13):
    data = np.array(boston.data[:, i])
    boston.data[:, i] = preprocessing.scale(data)
```

```python
# 変換後のデータ
print(boston.data[:, 0])
print(boston.data[:, 5])
```

### 次のセルに演習4-2の解答を記述してください

```python
# 演習4-2の解答欄

# PCA の読み込み
from sklearn.decomposition import PCA 
pca = PCA(n_components=2)
feature = pca.fit_transform(boston.data)
```

```python
# 次元削減後の変数と住宅価格の相関を確認。

for i in range(2):
    # 相関係数 
    coef = np.corrcoef(feature[:, i], boston.target)
    print('coef = ', coef[0,1])
    
    # グラフ化
    plt.xlabel(f'new_feature {i}')
    plt.ylabel('price')
    plt.scatter(feature[:, i], boston.target)
    plt.show()
```
